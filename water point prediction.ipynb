```python
X=scaled_df.drop(["Status"],axis=1)
y=scaled_df["Status"]
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=2323)

print(X_train.shape)

print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
trainpreds=lr.predict(X_train)
testpreds=lr.predict(X_test)
```

```python
from sklearn.metrics import  confusion_matrix,accuracy_score,recall_score,precision_score
print(confusion_matrix (y_train,trainpreds))   # accurate the predict 
print(confusion_matrix (y_test,testpreds))
print("train accuarcy:",accuracy_score(y_train,trainpreds))
print("test accuarcy:",accuracy_score(y_test,testpreds))
print("train recall:",recall_score(y_train,trainpreds))
print("test recall:",recall_score(y_test,testpreds))
print("train precision:",precision_score(y_train,trainpreds))
print("test precision:",precision_score(y_test,testpreds))
```

```python
from sklearn.tree import DecisionTreeClassifier
ds= DecisionTreeClassifier()
ds.fit(X_train,y_train)
```

```python
trainpreds_ds=ds.predict(X_train)
testpreds_ds=ds.predict(X_test)
```

```python
from sklearn.metrics import  confusion_matrix,accuracy_score,recall_score,precision_score
print(confusion_matrix (y_train,trainpreds_ds))   # accurate the predict 
print(confusion_matrix (y_test,testpreds_ds))
print("train accuarcy:",accuracy_score(y_train,trainpreds_ds))
print("test accuarcy:",accuracy_score(y_test,testpreds_ds))
print("train recall:",recall_score(y_train,trainpreds_ds))
print("test recall:",recall_score(y_test,testpreds_ds))
print("train precision:",precision_score(y_train,trainpreds_ds))
print("test precision:",precision_score(y_test,testpreds_ds))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
trainpreds_rf=rf.predict(X_train)
testpreds_rf=rf.predict(X_test)
```

```python
from sklearn.metrics import  confusion_matrix,accuracy_score,recall_score,precision_score
print(confusion_matrix (y_train,trainpreds_rf))   # accurate the predict 
print(confusion_matrix (y_test,testpreds_rf))
print("train accuarcy:",accuracy_score(y_train,trainpreds_rf))
print("test accuarcy:",accuracy_score(y_test,testpreds_rf))
print("train recall:",recall_score(y_train,trainpreds_rf))
print("test recall:",recall_score(y_test,testpreds_rf))
print("train precision:",precision_score(y_train,trainpreds_rf))
print("test precision:",precision_score(y_test,testpreds_rf))
```
